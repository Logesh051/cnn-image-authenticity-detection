```python
%%writefile app.py
import streamlit as st
import tensorflow as tf
import cv2
import numpy as np
from PIL import Image

# Load model from Google Drive
MODEL_PATH = "/content/drive/MyDrive/AI_Image_Project/ai_image_detector.keras"
model = tf.keras.models.load_model(MODEL_PATH, compile=False)

st.title("AI vs Real Image Detector")
st.write("Upload an image to check if it is AI-generated or real.")

uploaded_file = st.file_uploader(
    "Upload an image", type=["jpg", "jpeg", "png"]
)

def preprocess_image(image):
    # Force image to RGB (3 channels)
    image = image.convert("RGB")

    # Convert to numpy array
    img = np.array(image)

    # Resize to model input size
    img = cv2.resize(img, (224, 224))

    # Normalize pixel values
    img = img / 255.0

    # Add batch dimension
    img = np.expand_dims(img, axis=0)

    return img


if uploaded_file is not None:
    image = Image.open(uploaded_file)
    st.image(image, caption="Uploaded Image", use_column_width=True)

    img = preprocess_image(image)
    prediction = model.predict(img)[0][0]

    if prediction >= 0.5:
        st.success(f"AI-Generated Image ({prediction*100:.2f}%)")
    else:
        st.success(f"Real Image ({(1-prediction)*100:.2f}%)")
```
